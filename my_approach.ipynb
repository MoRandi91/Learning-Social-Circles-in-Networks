#### Считывание графа

```python
def read_nodeadjlist(filename):
    G = nx.Graph()
    for line in open(filename):
        e1, es = line.split(':')
        es = es.split()
        for e in es:
            if e == e1: continue
            G.add_edge(int(e1),int(e))
    return G
```

```python
def read_circles(filename):
    circles = []
    un_users = set()
    for line in open(filename):
        e1, es = line.split(':')
        es = es.split()
        circle = set()
        for e in es:
            circle.add(int(e))
            un_users.add(int(e))
        circles.append(circle)
    return circles, un_users
```

#### Алгоритмы

```python
# Compute the loss for one user (i.e., one set of circles for one ego-network)
# usersPerCircle: list of sets of users (groundtruth). Order doesn't matter.
# usersPerCircleP: list of sets of users (predicted). Order doesn't matter.
def loss1(usersPerCircle, usersPerCircleP):
  #psize: either the number of groundtruth, or the number of predicted circles (whichever is larger)
  psize = max(len(usersPerCircle),len(usersPerCircleP)) # Pad the matrix to be square
  # mm: matching matrix containing costs of matching groundtruth circles to predicted circles.
  #     mm[i][j] = cost of matching groundtruth circle i to predicted circle j
  mm = numpy.zeros((psize,psize))
  # mm2: copy of mm since the Munkres library destroys the data during computation
  mm2 = numpy.zeros((psize,psize))
  for i in range(psize):
    for j in range(psize):
      circleP = set() # Match to an empty circle (delete all users)
      circle = set() # Match to an empty circle (add all users)
      if (i < len(usersPerCircleP)):
        circleP = usersPerCircleP[i]
      if (j < len(usersPerCircle)):
        circle = usersPerCircle[j]
      nedits = len(circle.union(circleP)) - len(circle.intersection(circleP)) # Compute the edit distance between the two circles
      mm[i][j] = nedits
      mm2[i][j] = nedits

  if psize == 0:
    return 0 # Edge case in case there are no circles
  else:
    m = Munkres()
    #print mm2 # Print the pairwise cost matrix
    indices = m.compute(mm) # Compute the optimal alignment between predicted and groundtruth circles
    editCost = 0
    for row, column in indices:
      #print row,column # Print the optimal value selected by matching
      editCost += mm2[row][column]
    return int(editCost)
```

```python
def run_on_train(algorithm, power):
    ans = {}
    for userId in train_users:
        filename = str(userId) + '.egonet'
        G = read_nodeadjlist(egonets_dir + filename)

        circles = algorithm(G, power)
        ans[userId] = circles
    return ans
```

```python
# После работы алгоритма выкидываем все круги размера меньше или равные threshold

def filt(ans, threshold):
    f = {}
    for userId in train_users:
        circles = ans[userId]
        c = circles
        circles = filter(lambda circle: len(circle) > threshold, circles)
        if not circles:
            circles = filter(lambda circle: len(circle) == max(map(len,c)), c)
        f[userId] = circles
    return f
```

#### Первый простой алгоритм: Проходим по всем пользователям, если в каком-то круге он знает не менее чем power пользователей этого круга, то добавляем его в круг. Иначе создаем новый круг с этим пользователем.

```python
def my_algorithm(G, power):
     # find the single most popular user
    pageDictionary = nx.pagerank(G)
    pageRanks = list(pageDictionary.values())
    pageNodes = list(pageDictionary.keys())
    
    score = []
    for rank in pageRanks:
        score.append(rank)

    listOfSets = []
    for i, rank in enumerate(score):
        Ok = False
        for j in listOfSets:
            OK1 = True
            al = 0
            z = 0
            for k in j:
                if G.has_edge(k, pageNodes[i]):
                    z += 1
                al += 1

            OK1 = ((z + 0.0) / al) >= power
            if OK1:
                j.add(pageNodes[i])
                Ok = True
                break
        if Ok is False:
            newSets = set()
            newSets.add(pageNodes[i])
            listOfSets.append(newSets)
    return listOfSets
```

```python
th = range(5,15)
p = np.arange(1,10) / 10.0
res = np.zeros((len(th),len(p)))
for j, power in enumerate(p):
    ans = run_on_train(my_algorithm, power)
    for i, threshold in enumerate(th):
        my_ans = filt(ans, threshold)
        s = 0
        for userId in train_users:
            filename = circles_dir + str(userId) + '.circles'
            ans_circles, op = read_circles(filename)
            l = loss1(ans_circles, my_ans[userId])
            s+=l
            res[i,j] = (s/60.0*17)
    print j
```

### Алгоритм, который показал лучший результат на private: 7094, но не был выбран для финальной оценки.

##### Последовательно добавляем пользователи в круги согласно page-rank. Если пользователь знает не менее power в каком-то круге, то добавляем его в этот круг, иначе создаем новый. Затем фильтруем полученные круги, оставляя только тех, кто в итоге знает больше f_power пользователей круга. Затем оставляем круги, где число пользователей не менее minSizeCircle, при этом выкидываем круги, которые имеют мощность пересечения больше p_i с уже отобранными.

```python
def my_algorithm_best(G, power):
     # find the single most popular user
    f_power = 0.33
    minSizeCircle = 5
    p_i = 0.5
    
    pageDictionary = nx.pagerank(G)
    clustDictionary = nx.clustering(G)
    clustRanks = list(clustDictionary.values())
    clustNodes = list(clustDictionary.keys())

    score = []
    for i, rank in enumerate(clustRanks):
        score.append(pageDictionary.get(clustNodes[i]))


    sorted_score = {}
    for i, scor in enumerate(score):
        sorted_score[clustNodes[i]] = scor

    sorted_score = sorted(sorted_score.items(), key=operator.itemgetter(1), reverse=True)


    listOfSets = []
    for user, rank in sorted_score:
        Ok = False
        for j in listOfSets:
            OK1 = True
            al = 0
            z = 0
            for k in j:
                if G.has_edge(k, user):
                    z += 1
                al += 1

            OK1 = ((z + 0.0) / al) >= power
            if OK1:
                j.add(user)
                Ok = True
                break
        if Ok is False:
            newSets = set()
            newSets.add(user)
            listOfSets.append(newSets)

    ansListOfSets = []
    for circle in listOfSets:
        OK1 = True
        al = 0
        z = 0
        ans = set()
        for i in circle:
            for j in circle:
                if i == j: continue
                if G.has_edge(i, j):
                    z += 1
                al += 1
            OK1 = ((z + 0.0) / max(al, 1)) >= f_power
            if OK1:
                ans.add(i)
        if ans:
            ansListOfSets.append(ans)

    listOfSets = ansListOfSets

    listOfSets = sorted(listOfSets, key=lambda x: len(x), reverse=True)
    listOfCircles = []

    for community in listOfSets:
        # leave only relativly large communities
        #if len(community) >= max(5, G.number_of_nodes()*0.035):
        if len(community) >= minSizeCircle:
            OK = True
            ttt = 0
            for s in listOfCircles:
                for s1 in s:
                    for s2 in community:
                        if G.has_edge(s1, s2):
                            ttt += 1
                            break
                    if ttt >= p_i*len(community):
                        break
            if ttt < p_i*len(community):
                listOfCircles.append(list(community))

    if not listOfCircles:
        for community in listOfSets:
            # leave only relativly large communities
            if len(community) >= 10:
                OK = True
                ttt = 0
                for s in listOfCircles:
                    for s1 in s:
                        for s2 in community:
                            if G.has_edge(s1, s2):
                                ttt += 1
                                break
                        if ttt > 0.1*len(community):
                            break
                if ttt <= 0.1*len(community):
                    listOfCircles.append(list(community))

    if not listOfCircles:
        for community in listOfSets:
            # leave only relativly large communities
            if len(community) >= 8:
                OK = True
                ttt = 0
                for s in listOfCircles:
                    for s1 in s:
                        for s2 in community:
                            if G.has_edge(s1, s2):
                                ttt += 1
                                break
                        if ttt > 0.1*len(community):
                            break
                if ttt <= 0.1*len(community):
                    listOfCircles.append(list(community))

    if not listOfCircles:
        for community in listOfSets:
            # leave only relativly large communities
            if len(community) >= 6:
                OK = True
                ttt = 0
                for s in listOfCircles:
                    for s1 in s:
                        for s2 in community:
                            if G.has_edge(s1, s2):
                                ttt += 1
                                break
                        if ttt > 2:
                            break
                if ttt <= 2:
                    listOfCircles.append(list(community))


    if not listOfCircles:
        for community in listOfSets:
            # leave only relativly large communities
            if len(community) >= 5:
                OK = True
                ttt = 0
                for s in listOfCircles:
                    for s1 in s:
                        for s2 in community:
                            if G.has_edge(s1, s2):
                                ttt += 1
                                break
                        if ttt > 2:
                            break
                if ttt <= 2:
                    listOfCircles.append(list(community))


    if not listOfCircles:
        for community in listOfSets:
            # leave only relativly large communities
            if len(community) >= 4:
                OK = True
                ttt = 0
                for s in listOfCircles:
                    for s1 in s:
                        for s2 in community:
                            if G.has_edge(s1, s2):
                                ttt += 1
                                break
                        if ttt > 2:
                            break
                if ttt <= 2:
                    listOfCircles.append(list(community))

    listOfSets = listOfCircles
    return listOfSets
```
